```python

def visible_in_row(row, row_index):
    visible = set()
    tallest_left = int(row[0])
    tallest_right = int(row[len(row) - 1])
    for col_index, tree in enumerate(row):
        if int(tree) > tallest_left:
            visible.add(str(row_index) + '-' + str(col_index))
            tallest_left = int(tree)
    for i, tree in enumerate(row[::-1]):
        col_index = len(row) - 1 - i
        if int(tree) > tallest_right:
            visible.add(str(row_index) + '-' + str(col_index))
            tallest_right = int(tree)
    return visible
        
def visible_in_col(col, col_index):
    visible = set()
    tallest_left = int(col[0])
    tallest_right = int(col[len(col) - 1])
    for row_index, tree in enumerate(col):
        if int(tree) > tallest_left:
            visible.add(str(row_index) + '-' + str(col_index))
            tallest_left = int(tree)
    for i, tree in enumerate(col[::-1]):
        row_index = len(col) - 1 - i
        if int(tree) > tallest_right:
            visible.add(str(row_index) + '-' + str(col_index))
            tallest_right = int(tree)
    return visible
        

with open('input.txt', encoding="utf-8") as f:
    read_data = f.read()
    
    rows = read_data.splitlines()
    
    cols = [''] * len(rows)
    
    for row in rows:
        for col_index, item in enumerate(row):
            cols[col_index] += item
    
    visible = set()
    
    for row_index, row in enumerate(rows):
        visible.add(str(row_index) + '-' + '0')
        visible.add(str(row_index) + '-' + str(len(row) - 1))
        visible = visible.union(visible_in_row(row, row_index))
        
    for col_index, col in enumerate(cols):
        visible.add('0' + '-' + str(col_index))
        visible.add(str(len(col) - 1) + '-' + str(col_index))    
        visible = visible.union(visible_in_col(col, col_index))
        
    print(len(visible))
    
```

```python

def up_count(row_index, col_index, rows):
    tree = int(rows[row_index][col_index])
    count = 0
    row_index -= 1
    big_tree_found = False
    while row_index >= 0 and not big_tree_found:
        count += 1
        if int(rows[row_index][col_index]) >= tree:
            big_tree_found = True
        row_index -= 1
    return count

def down_count(row_index, col_index, rows):
    tree = int(rows[row_index][col_index])
    count = 0
    row_index += 1
    big_tree_found = False
    while row_index < len(rows) and not big_tree_found:
        count += 1
        if int(rows[row_index][col_index]) >= tree:
            big_tree_found = True
        row_index += 1
    return count

def left_count(row_index, col_index, rows):
    tree = int(rows[row_index][col_index])
    count = 0
    col_index -= 1
    big_tree_found = False
    while col_index >= 0 and not big_tree_found:
        count += 1
        if int(rows[row_index][col_index]) >= tree:
            big_tree_found = True
        col_index -= 1
    return count

def right_count(row_index, col_index, rows):
    tree = int(rows[row_index][col_index])
    count = 0
    col_index += 1
    big_tree_found = False
    while col_index < len(rows[row_index]) and not big_tree_found:
        count += 1
        if int(rows[row_index][col_index]) >= tree:
            big_tree_found = True
        col_index += 1
    return count

with open('input.txt', encoding="utf-8") as f:
    read_data = f.read()
    
    rows = read_data.splitlines()
    
    best_score = 0
    
    for row_index, row in enumerate(rows):
        for col_index, col in enumerate(row):
            up = up_count(row_index, col_index, rows)
            down = down_count(row_index, col_index, rows)
            left = left_count(row_index, col_index, rows)
            right = right_count(row_index, col_index, rows)
            
            tree_score = up * left * down * right
            
            best_score = max(best_score, tree_score)
    print(best_score)
            
    
    
```
